```python
def relu(x):
    return(np.maximum(0, x))
```

```python
w = np.array([-0.5, 2, -1])
x = np.array([1 , 2, 1.5])

s = w @ x
```

```python
relu(s)
```

```python
x = np.array([1, 1, 1])
w_1 = np.array([-0.5, 2, -1])
w_2 = np.array([3, 1.5, -1])

z_1 = w_1 @ x
z_2 = w_2 @ x

t_1 = relu(z_1)
t_2 = relu(z_2)
t_0 = 1

t = np.array([t_0, t_1, t_2])
w_3 = np.array([2, 3, -2])

h = w_3 @ t

y = sigmoid(h)
round(y, 2)
```
